```python
## create the simple txt file
import os
os.makedirs("../data/text_files",exist_ok=True)
```

```python
simple_text={
    "../data/text_files/test1.txt": "This is the content of test file 1."
}
for filepath,content in simple_text.items():
    with open(filepath,"w") as f:
        f.write(content)
print("Simple text files created successfully.")
```

```python
### text loader
# from langchain.document_loaders import TextLoader
from langchain_community.document_loaders import TextLoader
loader = TextLoader("../data/text_files/test1.txt",encoding="utf-8")
document = loader.load()
print (document)
```

```python
### directory loader
from langchain_community.document_loaders import DirectoryLoader,TextLoader
directory_loader = DirectoryLoader(
    "../data/text_files", 
    glob="*.txt", ## pattern to match only text files
    loader_cls=TextLoader, ## specify the loader class to use for each file
    loader_kwargs={"encoding": "utf-8"}, ## additional arguments for the loader
    show_progress=False
)
documents = directory_loader.load()
print(documents)
```

```python
from langchain_community.document_loaders import DirectoryLoader, PyMuPDFLoader

directory_loader = DirectoryLoader(
    "../data/pdf_files",
    glob="*.pdf",
    loader_cls=PyMuPDFLoader,
    show_progress=False
)

documents = directory_loader.load()

print(documents)
```
